```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = (10,6)
%config InlineBackend.figure_format = 'svg'
```

```python
x = np.arange(-2*np.pi,2*np.pi,0.1)
y = np.cos(x)
z = np.cos(2*x)
```

```python
plt.plot(x,y, label = 'y = cos(x)')
plt.plot(x,z, label = 'y = cos(2x)')
plt.legend()
plt.show()
```

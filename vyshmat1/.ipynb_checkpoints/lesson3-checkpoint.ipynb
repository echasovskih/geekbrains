```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#tema 2 zadanie 1
def vectlength(x,y,z):
    return np.sqrt(x**2+y**2+z**2)
```

```python
vectlength(10,10,10)
```

```python
#tema2 zadanie2
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)

fig, ax = plt.subplots()

ax.contour(x, y,(x**2 + y**2), [1], colors='k')
ax.grid(color='lightgray',linestyle='--')

ax.set_aspect(1)
plt.show()
```

```python
a = 4.
b = 2.

fig, ax = plt.subplots()

ax.contour(x, y,(x**2/a**2 + y**2/b**2), [1], colors='k')
ax.grid(color='lightgray',linestyle='--')

ax.set_aspect(1)
plt.show()

```

```python
a = 2.
b = 1.

fig, ax = plt.subplots()

ax.contour(x, y,(x**2/a**2 - y**2/b**2), [1], colors='k')
ax.grid(color='lightgray',linestyle='--')
ax.set_aspect(1)
plt.show()
```

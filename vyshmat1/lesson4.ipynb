```python
#zadanie3.1
import math as m
```

```python
def bern(n,k):
    return (1 / 2 ** n) * m.factorial(n) / (m.factorial(k) * m.factorial(n-k))
```

```python
bern(4,2)
```

```python
#zadanie3.2
bern(10,5)
```

```python
#tema 2 zadanie 1
def vectlength(x,y,z):
    return np.sqrt(x**2+y**2+z**2)
```

```python
#tema2 zadanie2
x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)

fig, ax = plt.subplots()

ax.contour(x, y,(x**2 + y**2), [1], colors='k')
ax.grid(color='lightgray',linestyle='--')

ax.set_aspect(1)
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)
```

```python
x1, y1 = fsolve(equations, (-2,0))
print(x1,y1)
```

```python
x1, y1 = fsolve(equations, (2,5))
print(x1,y1)
```

```python
x1, y1 = fsolve(equations, (4,15))
print(x1,y1)
```
